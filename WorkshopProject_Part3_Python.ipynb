# **Fractals and Multifractals**

```python
# Program_11a.py: The Koch Snowflake.
import numpy as np
import matplotlib.pyplot as plt
from math import floor
def Koch_Curve(k , angtri , xstart , ystart):
  n_lines , h = 4**k , 3**(-k)
  x, y, x[0],y[0] = [0]*(n_lines+1),[0]*(n_lines+1) , xstart , ystart
  segment=[0] * n_lines
  angle=[0, np.pi/3, -np.pi/3, 0]
  for i in range(n_lines):
    m , ang = i , angtri
    for j in range(k):
      segment[j] = np.mod(m, 4)
      m = floor(m / 4)
      ang = ang + angle[segment[j]]
    x[i+1] = x[i]+h*np.cos(ang)
    y[i+1] = y[i]+h*np.sin(ang)
  plt.axis("equal")
  plt.plot(x,y)
k = 6 # Plot the three sides.
Koch_Curve(k , np.pi/3 , 0 , 0)
Koch_Curve(k , -np.pi/3 , 0.5 , 0.866)
Koch_Curve(k , np.pi , 1 , 0)
plt.show()
```

```python
# Koch Curve
import numpy as np
import matplotlib.pyplot as plt
from math import floor
k = 2
n_lines = 4**k  #Number of Segments
h = 3**(-k)     #Length of Segments
x = [0] * (n_lines + 1)
y = [0] * (n_lines + 1)
x[0] , y[0] = 0 , 0
segment = [0] * n_lines
angle = [0 , np.pi / 3 , -np.pi / 3 , 0] # Recall pi randians = 180 degrees. moving it 60 degrees
for i in range (n_lines):
  m = i
  ang = 0
  for j in range(k):
    segment[j] = np.mod(m , 4) # 4 segments
    m =floor(m / 4)            # 4 segments
    ang = ang+ angle[segment[j]]
  x[i + 1] = x[i] + h * np.cos(ang)
  y[i + 1] = y[i] + h * np.sin(ang)
plt.axis("equal")
plt.plot(x , y)
plt.show()
```

```python
# Koch Square Fractral
import numpy as np
import matplotlib.pyplot as plt
from math import floor
k = 6
n_lines = 5**k  #Number of Segments
h = 3**(-k)     #Length of Segments
x = [0] * (n_lines + 1)
y = [0] * (n_lines + 1)
x[0] , y[0] = 0 , 0
segment = [0] * n_lines
angle = [0 , np.pi / 2 , 0 , -np.pi / 2 , 0] # Recall pi randians = 180 degrees. moving it 60 degrees
for i in range (n_lines):
  m = i
  ang = 0
  for j in range(k):
    segment[j] = np.mod(m , 5) #  segments
    m =floor(m / 5)            # 5 segments
    ang = ang+ angle[segment[j]]
  x[i + 1] = x[i] + h * np.cos(ang)
  y[i + 1] = y[i] + h * np.sin(ang)
plt.axis("equal")
plt.plot(x , y)
plt.show()
```

```python
# Multifractals Cantor set,  f-alpha curve
from sympy import log , symbols , diff
from sympy.plotting import plot_parametric
p1 , p2 = 1 / 9 , 8 /9 # p1+p2=1
q = symbols("q")
alpha = symbols("alpha")
tau = log(p1**q + p2**q) / log(3)
alpha = -diff(tau , q)
f = alpha * q + tau
plot_parametric(alpha , f , xlabel = r"$\alpha$" , ylabel = "f")
```
